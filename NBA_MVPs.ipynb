```python
import requests
import time
```

```python
dfs = []
for year in years:
    # opens dataframe file and reads it for every year
    with open(("mvp/{}.html".format(year)), encoding="utf8") as f:
        page = f.read()

    # creates parser class to extract table from page
    soup = BeautifulSoup(page, "html.parser")

    # removes unnecessary header from table
    # it finds 'tr' tag with class over_header and remove it from the HTML using decompose
    soup.find('tr', class_="over_header").decompose()

    # finds table with id MVP
    mvp_table = soup.find(id="mvp")

    # reads mvp HTML table as a string and creates column for specific year it came from
    mvp = pd.read_html(StringIO(str(mvp_table)))[0]
    mvp["Year"] = year

    # appends this mvp table into a dataframe
    dfs.append(mvp)
```

```python
# Combines all dataframes into 1 dataframe
mvps = pd.concat(dfs)
```

```python
mvps.tail()
```

```python
mvps.to_csv("mvps.csv")
```

```python
dfs = []
for year in years:
    with open(("player/{}.html".format(year)), encoding="utf8") as f:
        page = f.read()
        
    soup = BeautifulSoup(page, "html.parser")
    soup.find('tr', class_="thead").decompose()
    player_table = soup.find(id="per_game_stats")
    player = pd.read_html(StringIO(str(player_table)))[0]
    player["Year"] = year
    dfs.append(player)
```

```python
dfs = []
for year in years:
    with open(("team/{}.html".format(year)), encoding="utf8") as f:
        page = f.read()
    
    soup = BeautifulSoup(page, 'html.parser')
    remove_tag = soup.find_all('tr', class_="thead")
    for remove_tag in remove_tag:
        remove_tag.decompose()
        
    e_table = soup.find_all(id="divs_standings_E")[0]
    e_df = pd.read_html(StringIO(str(e_table)))[0]
    e_df["Year"] = year
    e_df["Team"] = e_df["Eastern Conference"]
    del e_df["Eastern Conference"]
    dfs.append(e_df)
    
    w_table = soup.find_all(id="divs_standings_W")[0]
    w_df = pd.read_html(StringIO(str(w_table)))[0]
    w_df["Year"] = year
    w_df["Team"] = w_df["Western Conference"]
    del w_df["Western Conference"]
    dfs.append(w_df)
```

```python
teams = pd.concat(dfs)
```
